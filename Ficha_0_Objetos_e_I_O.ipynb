# Problema 1 - Lei de Ohm


Escreva um programa que peça a tensão, $V$, e a corrente, $I$, num circuito e devolva a
tensão correspondente, com unidades.

$$R=\frac{V}{I}$$

```python
#@title Solução

v = float(input('Introduza a tensão em V: '))
i = float(input('Introduza a corrente em A: '))

print(f'A resistência no circuito é {v / i:.2f} Ohm.')
```

# Problema 2 - Energia Cinética

Escreva um programa que peça a massa, $m$, e a velocidade, $v$, de um objeto e calcule a energia cinética, com unidades.

$$E=\frac{1}{2}mv^2$$


```python
#@title Solução

m = float(input('Introduza a massa em kg: '))
v = float(input('Introduza a velocidade em m/s: '))

print(f'A energia cinética do objeto é {0.5 * m * v **2 :.2f} J.')
```

# Problema 3 - Fórmula Resolvente

Escreva um programa que determine as raízes para um polinómio de segundo grau do tipo $ax^2+bx+c=0$ através da Fórmula Resolvente:

$$x=\frac{-b\pm\sqrt{b^2-4ac}}{2a}$$

```python
#@title Solução

a = float(input('Introduza a: '))
b = float(input('Introduza b: '))
c = float(input('Introduza c: '))

x_pos = (-b+(b**2-4*a*c)**0.5)/(2*a)
x_neg = (-b-(b**2-4*a*c)**0.5)/(2*a)

print(f'A solução da equação {a}x^2+{b}x+{c}=0 são x={x_neg} ou x={x_pos}.')
```

# Problema 4 - Massas Molares

Escreva um programa que peça dois elementos químicos, o número de átomos e a sua massa atómica.

O programa deve devolver a fórmula química (compacta e expandida) e a massa molecular total arredondada às décimas.


```python
#@title Solução

a1 = input('Introduza o primeiro elemento: ')
n1 = int(input('Introduza o número de átomos: '))
m1 = float(input('Introduza a massa molar: '))

a2 = input('Introduza o segundo elemento: ')
n2 = int(input('Introduza o número de átomos: '))
m2 = float(input('Introduza a massa molar: '))

M = m1 * n1 + m2 * n2
molecula = a1 + str(n1) + a2 + str(n2)
molecula_exp = a1 * n1 + a2 * n2

print(f'A molécula {molecula} ({molecula_exp}) tem uma massa molar de {M :.2f} g/mol.')
```
