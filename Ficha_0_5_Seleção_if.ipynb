### Bilhete Piscina

Os preços de entrada numa piscina são:

- Crianças até 10 anos (inclusivé) não pagam;
- Crianças até aos 14 anos (inclusivé) pagam 3€;
- Jovens e adultos (14 aos 59 anos) pagam 5€;
- Pessoas a partir dos 60 anos (inclusivé) pagam 4€;
- A partir das 16:00 todos os valores reduzem 1€.

Escreva um programa que peça a hora de entrada e a idade de uma pessoa e calcule o valor que deverá pagar.

```python
#@title Solução

hora = int(input('Introduza a hora de entrada: '))
idade = int(input('Introduza a idade do cliente: '))

if idade <= 10:
  preco = 0
elif idade <= 14:
  preco = 3
elif idade <= 59:
  preco = 5
elif idade > 60:
  preco = 4

if hora >= 16 and preco !=0:
  preco = preco - 1

print(f'O valor a pagar é {preco}€')
```

### Pressão Arterial

A pressão arterial é caracterizada por dois valores:

- **Pressão arterial máxima ou sistólica:** valor durante a contração do coração (sístole);
- **Pressão arterial mínima ou diastólica:** valor durante o relaxamento do coração (diástole);

Para efeitos de diagnóstico e tratamento consideram-se as seguintes categorias:

- **Normal:** pressão sistólica ≤ 120 e pressão diastólica < 80;
- **Pré-hipertensão:** 120 < pressão sistólica < 130 e pressão diastólica > 80;
- **Hipertensão de nível 1:** 130 ≤ pressão sistólica < 140 ou 80 ≤ pressão diastólica < 90;
- **Hipertensão de nível 2:** 140 ≤ pressão sistólica < 180 ou pressão 90≤ pressão diastólica < 120;
- **Crise hipertensiva:** 180 ≤ pressão sistólica ou 120 ≤ pressão diastólica.

Escreva um programa que peça a pressão sistólica e diastólica de um paciente e imprima a categoria do paciente. Em caso de dúvida, considere sempre o caso mais crítico.

```python
#@title Solução

p_sistolica = float(input('Introduza a pressão arterial máxima (sistólica): '))
p_diastolica = float(input('Introduza a pressão arterial mínima (diastólica): '))

if p_sistolica >= 180 or p_diastolica >= 120:
  print('Paciente em crise hipertensiva.')
elif p_sistolica >= 140 or (p_diastolica >=90 and p_diastolica < 120):
  print('Paciente com hipertensão de nível 2.')
elif (p_sistolica >= 130 and p_sistolica <140) or (p_diastolica >= 80 and p_diastolica < 90):
  print('Paciente com hipertensão de nível 1.')
elif (p_sistolica > 120 and p_sistolica < 130) and (p_diastolica > 80):
  print('Paciente com pré-hipertensão.')
elif p_sistolica <= 129 and p_diastolica < 80:
  print('Paciente normal.')
```
